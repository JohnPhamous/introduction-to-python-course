## File Input

Variables are a great way to store information. The problem with variables however is that the data does not **persist**. What happens when you store data into a variable but restart your program?

We can save our data to files on our system so later on we can access them. Let's open up the terminal and create a sample text file.

```sh
touch text-input.txt # creates a new file
echo "Hello World!" >> text-input.txt # adds the string to the text file
cat text-input.txt # displays the content of the text file
```

Now we'll be able to use the contents of that text file inside our program.

````python
greeting = open('./text-input.txt')
print(greeting.read())
greeting.close()
````

`read()` allows us to read the all of the contents of a file into a single string. What if we wanted to separate the data?

Let's download this data file with stock data. Go back to your terminal and type in:

```sh
wget http://johnpham.net/cs190python/resources/NASDAQ100.txt
```

This file has the closing price per day going back to 1986. Let's create a program to find out what the largest delta was.

````python
data = open('./NASDAQ100.txt')
print(data.read())
````

There's a problem with the program above. It reads all of the data into 1 string. What if we wanted to separate it? Arrays are the perfect tool to do this.

````python
dataParsed = open('./NASDAQ100.txt')
print(dataParsed.readlines()[0].split(" ")[3])
dataParsed.close()
````

## File Output

File output lets us store the data that our program creates into files. Those files can then be used in our program later or in other programs.

````python
# (filename, mode)
outFile = open("bobo.txt", "w")
outFile.write("Hi this is Bobo\n")
outFile.close()

inFile = open("bobo.txt")
print(inFile.read())
inFile.close()

outFile = open("bobo.txt", "w")
outFile.write("I'm feeling kind of hungry\n")
outFile.close()

inFile = open("bobo.txt")
print(inFile.read())
inFile.close()

# What happened to the first thing we wrote?
````

## Using APIs with Requests

So far we've been downloading data through our terminal. That's an extra step that we don't need to do. We `requests`, we can download data directly from our program.

````python
import requests

rawData = requests.get('http://johnpham.net/cs190python/resources/NASDAQ100.txt')
print(rawData.text)

# This does not create a new file unlike before. The data is stored in your program's memory
````

If we wanted to, we can write the data to a file.

````python
dataFile = open('rawData.txt', 'w')
dataFile.write(rawData.text)
dataFile.close()

dataFile = open('rawData.txt')
print(dataFile.read())
dataFile.close()
````
